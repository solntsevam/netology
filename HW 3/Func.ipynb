## Инициализация переменных

```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

## Функции

```python
def find_doc(number):
    for el in documents:
        if el['number']==number:
            return el
```

```python
def document_owner(number):
    """
    Функция возвращает владельца документа по номеру документа
    """
    return find_doc(number)

def out_document_owner():
    """
    Функция запрашивает номер документа и выводит владельца документа на экран
    """
    number = input('Введите номер документа: ')
    owner  = document_owner(number)
    if owner != None:
        print('Владелец документа:',owner['name']) 
    else:
        print('Документ не найден в базе') 
```

```python
def document_dir(number):
    """
    Функция возвращает полку, на которой хранится документ по его номеру
    """
    for el in directories:
        if number in directories[el]:
            return(el) 

def out_document_dir():
    """
    Функция запрашивает номер документа и выводит полку, на которой он хранится
    """
    number = input('Введите номер документа: ')
    direct = document_dir(number)
    if direct != None:
        print('Документ хранится на полке: ', direct) 
    else:
        print('Документ не найден в базе') 
```

```python
def documents_info():
    """
    Функция выводит полную информацию по документам
    """
    print('Текущий список документов:')
    for el in documents:
        print(f"№ {el.get('number')}, тип: {el.get('type')}, владелец {el.get('name')}, полка хранения: {document_dir(el.get('number'))}")
```

```python
def new_dir():
    """
    Функция добавляет новую полку по номеру
    """
    while True:
        number = input('Введите номер новой полки: ')
        if number.isdigit():
            break
    if number in directories:
        print('Такая полка уже существует. Текущий перечень полок: '+ dir_info()+'.')
    else:
        directories.update({number:[]})
        print('Полка добавлена. Текущий перечень полок: '+ dir_info()+'.')
```

```python
def dir_info():
    """
    Функция возвращает список полок строкой c с разделителем ","
    """
    st = ''
    for el in directories:
        st = st+', '+el
    
    return str(st[2:])
```

```python
def del_dir():
    """
    Функция удаляет полку по номеру, если она пустая
    """
    while True:
        number = input('Введите номер новой полки: ')
        if number.isdigit():
            break
    if dir_exists(number):
        if len(directories[number]) != 0:
            print('На полке есть документы, удалите их перед удалением полки.')
        else:
            directories.pop(number)
            print('Полка удалена.')
    out_direct_info()
```

```python
def out_direct_info():
    print('Текущий перечень полок: '+ dir_info()+'.')
```

```python
def dir_exists(direct):
    if direct in directories:
        return True
    else:
        print('Такой полки не существует. Добавьте полку командой as.') 
        return False
```

```python
def add_doc():
    """
    Функция добавляет новый документ, и размещает его на полке
    """
    number   = input('Введите номер документа: ')
    type_doc = input('Введите тип документа: ')
    name = input('Введите владельца документа: ')
    direct = input('Введите полку для хранения: ')
    if dir_exists(direct):
        documents.append({'type': type_doc, 'number': number, 'name': name})
        directories[direct].append(number)
        print('Документ добавлен.')
    documents_info()    
```

```python
def del_doc():
    """
    Функция удаляет документ из базы
    """
    number   = input('Введите номер документа: ')
    doc = find_doc(number)
    if doc == None:
        print('Документ не найден в базе.')
    else:
        documents.remove(doc)
        del_doc_dir(number)
        print('Документ удален.')
    documents_info() 
```

```python
def del_doc_dir(number):
    """
    Функция удаляет документ с полки
    """
    doc_dir = directories[document_dir(number)]
    doc_dir.remove(number)
```

```python
def add_doc_dir(direct,number):
    """
    Функция добавляет документ на полку
    """
    directories[direct].append(number)
    
```

```python
def remove_doc():
    """
    Функция перекладывает документ на другую полку
    """
    number   = input('Введите номер документа: ')
    direct   = input('Введите полку для хранения: ')
    if dir_exists(direct):
        del_doc_dir(number)
        add_doc_dir(direct,number)
        print('Документ перемещен.')
    documents_info()
        
```

## Тело программы

```python
while True:
    res = input('Введите команду (q-для выхода):').upper()
    if res == 'Q':                  # выход
        break
    elif res == 'P':                # вывод владельца по номеру
        out_document_owner() 
    elif res == 'S':                # вывод полки хранения по номеру документа
        out_document_dir()
    elif res == 'L':                # вывод полной информации по документам
        documents_info()
    elif res == 'AS':               # добавить новую полку
        new_dir()
    elif res == 'DS':               # удалить полку
        del_dir()
    elif res == 'AD':               # добавить новый документ, разместить его на полке
        add_doc() 
    elif res == 'D':                # удалить документ
        del_doc()
    elif res == 'M':                # переложить документ на другую
        remove_doc()
    else:
        print('Неизвестная команда')
    
    
print('Работа с программой завершена. Удачного дня')
```

```python
stream = [
'2018-01-01,user100,150',
'2018-01-07,user99,205',
'2018-03-29,user1001,81'
]
```

```python
UnUser = []
summa = 0.0
for el in stream:
    el = el.split(',')
    if el[1] not in UnUser:
        UnUser.append(el[1])
    summa = summa+int(el[2])
avr = summa / len(UnUser)
print(round(avr,2))
```
